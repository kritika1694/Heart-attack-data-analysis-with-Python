# 5. Visualisations

```python
num_df = df.select_dtypes(include=['number'])
num_df.corr()                                        # correlation analysis of numeric variables
```

```python
sns.heatmap(num_df.corr(), annot= True)
plt.rcParams['figure.figsize'] =(10,5)
plt.title('Correlation matrix of numeric columns')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8, 5))                                      #histogram plot in seaborn for numeric columns
sns.histplot(data=df, x='Age', bins=30, color='#a3c2e8')  


plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')


plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Blood Pressure (mmHg)', bins=30, color='#f4b5a6')  

plt.title('Blood Pressure (mmHg) Distribution')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Cholesterol (mg/dL)', bins=30, color='#d4e8a4')

plt.title('Cholesterol (mg/dL) Distribution')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Frequency')                 
plt.show()
```

```python
object_columns= ['Gender','Has Diabetes', 'Smoking Status', 'Chest Pain Type', 'Treatment']
```

```python
for col in object_columns:                         # Loop through categorical column
   
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, hue = 'Gender', palette='pastel')   # Create a bar plot using Seaborn

   
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
```
